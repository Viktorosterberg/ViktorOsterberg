# Task 1: A list of dictionaries

```python
transactions = [
    {"id": 1, "product": "Laptop",   "price": 1200, "quantity":  2, "region": "EU"},
    {"id": 2, "product": "Phone",    "price":  800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet",   "price":    0, "quantity":  5, "region": "EU"},
    {"id": 4, "product": "Monitor",  "price":  350, "quantity":  3, "region": "US"},
    {"id": 5, "product": "Laptop",   "price": 1200, "quantity":  1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price":   75, "quantity": 10, "region": "APAC"},
]
 
# a) Filter out invalid transactions (price == 0 or quantity < 0)
valid_transactions = [
    t for t in transactions if t["price"] > 0 and t["quantity"] >= 0
]
 
# b) Add a 'total' field to each valid transaction
for t in valid_transactions:
    t["total"] = t["price"] * t["quantity"]
 
# c) Group by region and sum revenue
revenue_by_region = {}
for t in valid_transactions:
    region = t["region"]
    revenue_by_region[region] = revenue_by_region.get(region, 0) + t["total"]
    
print("Revenue by region:", revenue_by_region)

# d) Print the region with the highest total revenue
top_region = max(revenue_by_region, key=revenue_by_region.get)
print(f"Top region: {top_region} with revenue {revenue_by_region[top_region]}")
```

# Task 2: Creating a function

```python
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]

def summarise_sales(s):
    total   = sum(s)
    average = round(total / len(s), 2)
    return {
        "total":               total,
        "average":             average,
        "best_day":            s.index(max(s)) + 1,
        "worst_day":           s.index(min(s)) + 1,
        "above_average_count": sum(1 for x in s if x > average),
    }

print(summarise_sales(sales))
```

# Task 3: Loops

```python
students = [
    {"name": "Anna", "grade": 85, "attendance": 0.92},
    {"name": "Erik", "grade": 58, "attendance": 0.75},
    {"name": "Mia",  "grade": 91, "attendance": 0.98},
    {"name": "Leo",  "grade": 44, "attendance": 0.60},
    {"name": "Sara", "grade": 72, "attendance": 0.88},
    {"name": "Tom",  "grade": 67, "attendance": 0.70},
    {"name": "Nina", "grade": 95, "attendance": 0.99},
    {"name": "Olli", "grade": 50, "attendance": 0.65},
]
 
# a) Students with grade >= 50
passing = [s["name"] for s in students if s["grade"] >= 50]
print("Passing students:", passing)
 
# b) Status for each student
print()
for s in students:
    if s["grade"] >= 80:
        status = "Excellent"
    elif s["grade"] >= 50:
        status = "Pass"
    else:
        status = "Fail"
    print(f"{s['name']}: {status}")
 
# c) The count of students with attendance below 80%
low_attendance_count = sum(1 for s in students if s["attendance"] < 0.80)
print(f"\nStudents with attendance below 80%: {low_attendance_count}")
 
# d) Highest grade using a manual (loop)
top_student = students[0]
for s in students[1:]:
    if s["grade"] > top_student["grade"]:
        top_student = s
print(f"\nTop student (loop): {top_student['name']}  {top_student['grade']}")
 
# e) Highest grade using max
top_student_max = max(students, key=lambda s: s["grade"])
print(f"Top student (max):  {top_student_max['name']}  {top_student_max['grade']}")
```

# Task 4: Class and methods

```python
class Portfolio:
    def __init__(self, owner):
        self.owner    = owner
        self.holdings = {}
 
    def add_asset(self, ticker, shares, price_per_share):
        """Add a new asset or update shares/price if ticker already exists."""
        if ticker in self.holdings:
            self.holdings[ticker]["shares"] += shares
        else:
            self.holdings[ticker] = {"shares": shares}
        self.holdings[ticker]["price_per_share"] = price_per_share
 
    def total_value(self):
        """Return the total market value of the portfolio."""
        return sum(
            data["shares"] * data["price_per_share"]
            for data in self.holdings.values())
 
    def most_valuable(self):
        """Return the ticker with the highest total value."""
        return max(
            self.holdings,
            key=lambda t: self.holdings[t]["shares"] * self.holdings[t]["price_per_share"])
 
    def __str__(self):
        return (
            f"Portfolio of {self.owner} | "
            f"Assets: {len(self.holdings)} | "
            f"Total value: €{self.total_value():,.2f}")
 
# Portfolio
p = Portfolio("Viktor")
p.add_asset("NVDA",  12, 1000)
p.add_asset("AMZN",   5, 400)
p.add_asset("GOOGL",  14, 500)
p.add_asset("NVDA",   11, 1000)
 
print(p)
print(f"Total value:   €{p.total_value():,.2f}")
print(f"Most valuable: {p.most_valuable()}")
```

```python
from datetime import date
 
orders = [
    ("2025-01-15", "Laptop"),  ("2025-01-22", "Phone"),
    ("2025-02-03", "Laptop"),  ("2025-02-14", "Tablet"),
    ("2025-03-01", "Phone"),   ("2025-03-10", "Phone"),
    ("2025-03-22", "Laptop"),  ("2025-04-05", "Monitor"),
]
 
# Parse each date string into a date object
parsed_dates = [date.fromisoformat(order[0]) for order in orders]
 
earliest = min(parsed_dates)
latest   = max(parsed_dates)
delta    = (latest - earliest).days
 
print(f"Earliest order: {earliest}")
print(f"Latest order:   {latest}")
print(f"Days between:   {delta}")
```

# Task 7: Pandas

```python
import pandas as pd

data = {
    "employee":          ["Anna", "Erik", "Mia", "Leo", "Sara", "Tom"],
    "department":        ["Finance", "Marketing", "Finance", "IT", "Marketing", "IT"],
    "salary":            [4500, 4000, 4800, 5200, 3900, 5100],
    "years_exp":         [5, 3, 7, 10, 2, 8],
    "performance_score": [4.2, 3.8, 4.5, 4.0, 3.5, 4.9],
}
df = pd.DataFrame(data)
 
# Employee number as index (starting from 1)
df.index = range(1, len(df) + 1)
df.index.name = "employee_no"
 
# a) Add salary_per_year_exp column
df["salary_per_year_exp"] = (df["salary"] / df["years_exp"]).round(0).astype(int)
 
print("Part A: Table with new column 'salary_per_year_exp'")
print(df.to_string())
print()
 
# b) Top 3 employees by performance score
top3 = df.sort_values("performance_score", ascending=False).head(3)
 
print("Part B: Top 3 employees by performance score")
print(top3[["employee", "performance_score"]].to_string())
print()
 
# c) Finance employees with performance score > 4.0
finance_high = df.loc[(df["department"] == "Finance") & (df["performance_score"] > 4.0)]
 
print("Part C: Finance employees with performance score > 4.0")
print(finance_high.to_string())
print()

# d) Rename years_exp -> experience_years
df = df.rename(columns={"years_exp": "experience_years"})
 
print("Part D: Table after renaming 'years_exp' to 'experience_years'")
print(df.to_string())
print()
 
# e) Salary & performance score for first 4 employees
print("Part E: Salary & performance score for first 4 employees (iloc)")
print(df.iloc[:4][["salary", "performance_score"]].to_string())
```

# Task 8: More Pandas

```python
import numpy as np
import pandas as pd

rng             = np.random.default_rng(seed=123)
days            = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors        = rng.integers(200, 600, size=28)
conversion_rate = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value = rng.normal(50, 10, size=28).round(2)
 
# a) Create DataFrame with date as index
traffic = pd.DataFrame({
    "visitors":        visitors,
    "conversion_rate": conversion_rate,
    "avg_order_value": avg_order_value,
}, index=days)
traffic.index.name = "date"
 
# b) Add new columns
 
# conversions = visitors × conversion_rate
traffic["conversions"] = (traffic["visitors"] * traffic["conversion_rate"]).round(0).astype(int)
 
# daily_revenue = conversions × avg_order_value
traffic["daily_revenue"] = (traffic["conversions"] * traffic["avg_order_value"]).round(2)
 
print(traffic.to_string())
print(f"\nTotal revenue over 4 weeks: €{traffic['daily_revenue'].sum():,.2f}")
 
```
